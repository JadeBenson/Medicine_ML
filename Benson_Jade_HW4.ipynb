```python
#function to read in the text files and clean them 

def read_clean_txt(file_directory):
    """
    Reads in the text from the files and prints out lines 
    Cleans the set-up and empty lines
    
    inputs
    file_directory: the string file location 
    
    outputs 
    cleaned_list: nested list of lines with relevant information
    
    """
    response_list = []
    with open(file_directory) as f:
        response_list = f.readlines()
    
    #new list to store the cleaned text
    cleaned_list = []

    #from earlier testing the GPT set up lasts 613 lines 
    for line in response_list[613:]: 
        #if line is not empty 
        if line != ' \n' and line != '\n':
            #and line is not the GPT-J asking for a prompt 
            if line != 'Enter prompt or quit: Setting `pad_token_id` to `eos_token_id`:50256 for open-end generation.\n':
                #then add it to cleaned list
                cleaned_list.append(line)
    
    return cleaned_list
    
```

2. Does GPT-J consistently predict the same drug target across different prompts (mechanism of action, inhibits, targets)? 

We will select 100 drugs from the first text file and cross-reference the responses across the three looking for similar keywords. 

```python
random.seed(1234)

drugs = []
inhibits_responses = random.choices(drug_inhibits_responses, k = 100)

for response in inhibits_responses:
    drug_is = response.split()[1]
    #get ride of "is" at the end of the drug name
    drug = drug_is[:len(drug_is)-2]
    drugs.append(drug)
    
```

```python
mechanisms_responses = []
for drug in drugs: 
    for line in drug_mechanism_responses: 
        if drug in line: 
            mechanisms_responses.append(line)
```

```python
target_responses = []
for drug in drugs: 
    for line in drug_target_responses: 
        if drug in line: 
            target_responses.append(line)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('punkt', download_dir='./nltk_data') 
nltk.download('stopwords', download_dir='./nltk_data') 
from nltk.tokenize import word_tokenize
```

```python
import string

stop_tokens = ["response", "drug",  "."] + stopwords.words('english') + word_tokenize(string.punctuation)

def getResponse(collection, substring):
    response = [x.lower() for x in collection if substring in x][0]
    response_tokens = word_tokenize(response)
    #return string without stop words 
    return [word for word in response_tokens if not word in stop_tokens]
```

```python
def calculateSimilarity(response1, response2, response3):
    maximum_response = max(response1, response2, response3)
    similar_words = [word for word in maximum_response if word in response1 and word in response2 and word in response3]
    return [similar_words, len(similar_words)/len(maximum_response)]
    
```

```python
for drug in drugs:
    inhibit = getResponse(inhibits_responses, drug)
    #print(inhibit)
    mechanism = getResponse(mechanisms_responses, drug)
    #print(mechanism)
    target = getResponse(target_responses, drug)
    #print(target)
    similarities = calculateSimilarity(inhibit, mechanism, target)
    print("Drug " + drug + " has a meaningful word matching proportion of: " , similarities[1])
    print("Drug " + drug + " has these similar words appear across queries: " , similarities[0])
        
```

This is really bad. Most of these don't have any words in common across the three responses. If they do, it's the drug name or some lingering stop words or verbs that should be cleaned. This suggests that the accuracy is not high across the responses or they might be reporting on different aspects of drugs.

In this problem, we will extract the gene name and product from the product output text file, see if that occurs in the drug targets text file. Then if a drug target occurs, we search to see if it has a patent associated with it. 

First we will find gene symbols, names and products for the table and put them into separate lists to be wrapped into dataframe later

```python
gene_symbols = []
gene_names = []
gene_products = []

for line in gene_products_responses[1:]: 
    split = line.split()
    
    gene_symbol = split[4]
    gene_symbols.append(gene_symbol)
        
    end_index = split.index("gene.")
    gene_name = ' '.join(split[7:end_index])
    gene_names.append(gene_name)
        
    product_split = "The gene product of " + gene_symbol + " is "
    brief_split = line.split(product_split,1)[1]
    
    #only want brief description
    gene_product = brief_split.split(".", 1)[0]
    if len(gene_product)>0: 
        gene_products.append(gene_product)
    else: 
        gene_products.append(None)
        
                
    
        
```

```python
print(len(gene_symbols))
print(len(gene_names))
print(len(gene_products))
```

We then look whether these genes have any matching drugs in the drug targets output

```python
drug_names = []
found = False
for gs in gene_symbols: 
    found = False
    for line in drug_target_responses:
        if gs in line:
            drug_names.append(line.split()[1])
            found = True
            break
    if found == False:
        drug_names.append(None)
                
```

```python
print(len(drug_names))
```

Lastly, we see if there are any reported patents for the drugs. According to the US Patent office, patent numbers can only be 6, 7 or 8 digits (https://www.uspto.gov/patents/apply/applying-online/patent-number). We will look for any of these and include them. 

```python
import re

patent_numbers = []
found = False

for d in drug_names: 
    if d == None:
        patent_numbers.append(None)
        
    else:
        found = False
        for line in patent_number_responses:
            if d in line:
                #6
                six_digits = re.findall(r"\D(\d{6})\D", line)
                    
                #7 
                seven_digits = re.findall(r"\D(\d{7})\D", line)
                    
                #8 
                eight_digits = re.findall(r"\D(\d{8})\D", line)

                possible_patents = six_digits + seven_digits + eight_digits
                    
                if len(possible_patents)>0: 
                    patent_numbers.append(possible_patents)
                    
                else: 
                    patent_numbers.append(None)
                    
                found = True 
                break
                
        if found == False: 
            patent_numbers.append(None)
        
   
```

```python
print(len(patent_numbers))
```

```python
import pandas as pd 

linked_table = pd.DataFrame()
linked_table['Gene_Symbol'] = gene_symbols
linked_table['Gene_Name'] = gene_names
linked_table['Gene_Product'] = gene_products
linked_table['Drug_Name'] = drug_names
linked_table['Patent_Numbers'] = patent_numbers

#some issues with the gene product
#perhaps to be expected given its changing formating 
linked_table.head()
```
